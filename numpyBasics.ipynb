Compare list and numpy array 

```python
import time

n= 1_000_000

py_list = list(range(n))
np_array  = np.arange(n)


print(len(py_list))
print(len(np_array))

print(py_list[:10])
print(np_array[:10])

start = time.time()
py_res = [x*2 for x in py_list] 
end = time.time()
print(f"Python list took {end - start:.5f} seconds")

start = time.time()
np_res = np_array * 2   
end = time.time()
print(f"Numpy array took {end - start:.5f} seconds")
```
