Compare list and numpy array 

```python
import time

n= 1_000_000

py_list = list(range(n))
np_array  = np.arange(n)


print(len(py_list))
print(len(np_array))

print(py_list[:10])
print(np_array[:10])

start = time.time()
py_res = [x*2 for x in py_list] 
end = time.time()
print(f"Python list took {end - start:.5f} seconds")

start = time.time()
np_res = np_array * 2   
end = time.time()
print(f"Numpy array took {end - start:.5f} seconds")
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([[4, 5, 6], [4,5,6]])
arr3 = np.array([[[4, 5, 6], [4,5,6]], [[4, 5, 6], [4,5,6]]])

print('\n1D Array:\n', arr1)
print(type(arr1))
print('\nShape of arr1:\n', arr1.shape)
print('\n2D Array\n', arr2)
print(type(arr2))
print('\nShape of arr2:\n', arr2.shape)
```

Array Attributes

```python
print('\n3D Array\n', arr3)

print('Shape of arr1:', arr1.shape)
print('Shape of arr2:', arr2.shape)
print('Shape of arr3:', arr3.shape)

print('\n')

print('Number of dimensions of arr1:', arr1.ndim)
print('Number of dimensions of arr2:', arr2.ndim)  
print('Number of dimensions of arr3:', arr3.ndim)

print('\n')

print('Data type of arr1:', arr1.dtype)
print('Data type of arr2:', arr2.dtype)
print('Data type of arr3:', arr3.dtype)

print('Number of elements in arr1:', arr1.size)
print('Number of elements in arr2:', arr2.size)
print('Number of elements in arr3:', arr3.size)
```

**Creating Special Arrays**

```python
zeros = np.zeros((3, 3))
print('\n 3x3 array of zeros:\n', zeros)

ones = np.ones((4, 4))
print('\n 4x4 array of ones:\n', ones)

random_array = np.random.rand(3, 3)
print('\n 3x3 array of random numbers between 0 and 1:\n', random_array)

eye_matrix = np.eye(3)
print('\n 4x4 identity matrix:\n', eye_matrix)

array_range = np.arange(0, 10, 2)
print('\n Array of even numbers from 0 to 10:\n', array_range)

```
