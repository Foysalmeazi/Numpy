Compare list and numpy array 

```python
import time

n= 1_000_000

py_list = list(range(n))
np_array  = np.arange(n)


print(len(py_list))
print(len(np_array))

print(py_list[:10])
print(np_array[:10])

start = time.time()
py_res = [x*2 for x in py_list] 
end = time.time()
print(f"Python list took {end - start:.5f} seconds")

start = time.time()
np_res = np_array * 2   
end = time.time()
print(f"Numpy array took {end - start:.5f} seconds")
```

**Creating Special Arrays**

```python
zeros = np.zeros((3, 3))
print('\n 3x3 array of zeros:\n', zeros)

ones = np.ones((4, 4))
print('\n 4x4 array of ones:\n', ones)

random_array = np.random.rand(3, 3)
print('\n 3x3 array of random numbers between 0 and 1:\n', random_array)

eye_matrix = np.eye(3)
print('\n 4x4 identity matrix:\n', eye_matrix)

```

```python
# Example: Matrix multiplication with identity and zero matrices

import numpy as np

# Create a sample 2x2 matrix
A = np.array([[1, 2], [3, 4]])
print("Original matrix A:")
print(A)
print()

# Identity matrix multiplication (preserves A)
I = np.eye(2)
result_I = A @ I  # or np.dot(A, I)
print("A multiplied by identity matrix I:")
print(result_I)
print("Notice: A @ I = A (unchanged)")
print()

# Zero matrix multiplication (changes to all zeros)
Z = np.zeros((2, 2))
result_Z = A @ Z  # or np.dot(A, Z)
print("A multiplied by zero matrix Z:")
print(result_Z)
print("Notice: A @ Z = all zeros (changed)")
```
